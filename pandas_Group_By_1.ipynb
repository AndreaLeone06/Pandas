```python
film=pd.read_csv("https://raw.githubusercontent.com/wtitze/3E/main/2010.csv",delimiter=";")
```

```python
film
```

1. il numero di film per ogni lingua. Ordinare in ordine decrescente sul numero di film

```python
film.groupby("Language")[["Title"]].count().sort_values(by=["Title"],ascending=False).reset_index()
```

```python
film2=film.groupby("Language").count().sort_values(by=["Title"],ascending=False).reset_index()
film2[["Language","Title"]]
```

2. il numero di film per ogni nazione.

```python
film.groupby("Country")[["Country"]].count()
```

3. il budget totale che ogni nazione ha investito nei film. Ordinare in ordine decrescente base al budget totale

```python
film.groupby("Country")["Budget"].sum().reset_index().sort_values(by=["Budget"],ascending=False)
```

4. la nazione che ha investito il budget più alto

```python
filmgroup=film.groupby("Country")["Budget"].sum().reset_index().sort_values("Budget").reset_index()
```

```python
filmgroup
```

```python
filmgroup[-1:]
```

5. il budget medio che ogni nazione ha investito nei film. Ordinare in ordine decrescente base al budget medio

```python
film.groupby("Country")["Budget"].mean().reset_index().sort_values("Budget",ascending=False)
```

6. le nazioni che hanno investito, in media, un budget più basso rispetto alla media complessiva dei budget delle nazioni

```python
film3=film.groupby("Country")["Budget"].mean().reset_index()
```

```python
film3
```

```python
film3[film3.Budget < film.Budget.mean()]
```
