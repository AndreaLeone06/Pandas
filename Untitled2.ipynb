visualizzare 3 province casuali non ripetute

```python
prima = random.randint(0, 11)
seconda = random.randint(0, 11)
terza = random.randint(0, 11)
if prima == seconda or seconda == terza or prima == terza:
    prima = province[random.randint(0, 12)]
    seconda = province[random.randint(0, 12)]
    terza = province[random.randint(0, 12)]
else:
  print(province[prima])
  print(province[seconda])
  print(province[terza])
```

modificare la lista trasformando monza e brianza in 2 province separate

```python
pos= province.index("Monza e della Brianza")
for indice in range(len(province)):
  if province[indice] == "Monza e della Brianza":
    province.pop(pos)
    province[pos] == "Monza"
    province[pos + 1] == "Brianza"
print(province)
```

esercizio 2

caricare una lista con i primi 50 multipli di 5

```python
lista1 = [indice for indice in range(5,255) if indice%5==0]
print(lista1)
```

caricare in un'altra lista i multipli di 10

```python
lista2 = []
for indice in range(len(lista1)):
  if lista1[indice] % 10==0:
    lista2.append(lista1[indice])
print(lista2)
```

eliminare nella seconda lista tutti i valori compresi tra 30 e 60; visualizzare il numero di elementi nella lista e ilnumero di qualli eliminati.

```python
conto = 0
for indice in range(len(lista2)):
  if lista2[indice] >= 30 <= 60:
    conto += 1
    del lista2[indice]
print(len(lista2), conto, lista2)
```
