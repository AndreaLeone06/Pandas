```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/platea.csv")
```

```python
df
```

il numero di record presente nel file

```python
len(df)
```

il numero di regioni presente nel file

```python
df.groupby("reg")[["eta"]].count()
```

 il numero di abitanti di ogni regione. Ordinare in ordine decrescente sul numero di abitanti


```python
dfbar= df.groupby("reg")["totale_popolazione"].sum().reset_index().sort_values(by=["totale_popolazione"],ascending=False)
dfbar
```

la regione con il numero di abitanti piÃ¹ alto

```python
dfbar[:1].reset_index()
```

```python
x= df.groupby("reg")["totale_popolazione"].sum().max()
x
```

```python
dfbar[dfbar.totale_popolazione == x][['reg']]
```

 il numero medio di abitanti delle regioni

```python
max=df.groupby("reg")["totale_popolazione"].sum().mean()
max
```

 le regioni che hanno piÃ¹ abitanti della media degli abitanti delle regioni

```python
dfbar[dfbar.totale_popolazione > max]
```

 la regione che ha piÃ¹ abitanti nella fascia 12-19

```python
df.groupby(df.reg[df.eta=="12-19"])[["totale_popolazione"]].sum().sort_values(by=["totale_popolazione"],ascending=False)[:1]
```

```python
y= df.groupby(df.reg[df.eta=="12-19"])[["totale_popolazione"]].sum().max()
y
```

```python
df[df.reg[df.eta=="12-19"]["totale_popolazione"] == y][['reg']]
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.barh(dfbar.reg,dfbar.totale_popolazione,label="abitanti")
plt.yticks(fontsize=15)
ax.legend()
ax.set_ylabel('regioni')
ax.set_xlabel('abitanti')
ax.set_title('numero di abitanti di ogni regione')
```
