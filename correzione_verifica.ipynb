caricare una lista con i primi 50 multipli di 5

```python
lista1 = [indice for indice in range(5,255,5)]
print(lista1)
```

caricare in un'altra lista i multipli di 10

```python
lista2 = []
for indice in range(len(lista1)):
  if lista1[indice] % 10==0:
    lista2.append(lista1[indice])
print(lista2)
```

```python
lista2 = [mul for mul in lista1 if mul%10==0]
print(lista2)
```

eliminare dalla seconda lista tutti i valori compresi tra 30 e 60

```python
count=0
for valore in lista2:
  if valore >= 30 and valore <=160:
     lista2.remove(valore)
  count+=1
print(lista2)
```

caricare in una nupova lista i cinque elementi centrali

```python
multiplicentrali=[]
multiplicentrali=lista2[len(lista2)//2-2:len(lista2)//2+3]
print(multiplicentrali)
```

moltiplicare per 10 tutti i numeri della lista precedente


```python
prodotti=[numero*10 for numero in multiplicentrali]
print(prodotti)
```

mettere in un'unica lista tutti i valori

```python
listatutto=lista1+lista2+multiplicentrali+prodotti
print(listatutto)
```
