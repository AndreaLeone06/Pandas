caricare una lista con i primi 50 multipli di 5

```python
lista1 = [indice for indice in range(250) if indice%5==0]
lista1.pop(0)
print(lista1)
```

caricare in un'altra lista i multipli di 10

```python
lista2 = []
for indice in range(len(lista1)):
  if lista1[indice] % 10==0:
    lista2.append(lista1[indice])
print(lista2)
```

eliminare nella seconda lista tutti i valori compresi tra 30 e 60; visualizzare il numero di elementi nella lista e ilnumero di qualli eliminati.

```python
conto = 0
for indice in range(len(lista2)):
  if lista2[indice] >= 30 <= 60:
    conto += 1
    del lista2[indice]
print(len(lista2), conto, lista2)
```
